```python
a = "Python Hello"

a = a[7:] + ' ' + a[:6]
a
```

문자열의 값이 변화시킬 수 있다면 좋겠습니다만, 그렇지 않아서 아예 새로 대입을 합니다.
'Hello'가 시작되는 시점인 7번째부터 끝까지를 하나로 엮어 'Hello'를 전반부에, 'Python'의 마지막 시점인 제 6 위치 이전까지를 하나로 엮어 후반부에 넣고 그 사이에 공백을 직접 대입했습니다.

```python
b = "Hello Python World"
b = b[-5:] + b[5:-5] + b[:5]
b
```

위와 굉장히 유사합니다. 차이점이라면 사이에 있는 부분을 그 사이의 공백을 포함해서 잘라내어, 위에서처럼 따로 공백을 더해주지 않고도 지정된 과제가 수행 가능했습니다.

```python
c = "Hello"
c = c[::-1]
c
```

이것에 대한 구체적 원리에 대한 설명은 슬라이싱의 고찰로 이어집니다. 자세한 것은 4번의 문제에서도 언급하겠지만 위 구문의 효과는 c의 내용을 거꾸로 해서 만든 값에 c를 새롭게 배정해 주는 것이죠. 물론 굳이 저렇게 하지 않아도 reverse 함수를 사용한다면 문제 없이 할 수 있지만 일단은 이런 방법으로 해 보았습니다.

```python
s = "python"
print s[0], s[0][0], s[0][0][0]
```

이 문제와 다음 문제는 '정확하게 설명' 이기 때문에 파이썬의 문자열 슬라이싱에 대한 확인 결과를 설명해두고 넘어가려 합니다.

#*문자열 이름*  [ 출발칸 : 정지칸 : 이동 간격 ]

이 위치별 역할이고, 이 결과로 반환되는 것은
1. 문자열 이름에 해당하는 레퍼런스가 가지는 문자열을 확인해서 도로에 깐다.
2. 출발 칸에서 시작해 정지칸에 도착하기 전까지 이동하는 칸들의 값을 읽는다.
3. 이동하는 것은 이동 간격에 쓰인 칸만큼으로 한다.
4. 만약 이동하는 것이 의미가 없다면 그냥 제자리에서 텅 빈 문자열을 리턴한다.

와 같은 식이라고 생각하는데, 이를 C 언어 식의 유사 코드로 표현하자면 이렇게 될 겁니다.

tmp = ''
for(i=출발; i < 도착; i += 이동)

    if((이동 방향과 출발 - 도착 사이 방향이 같지 않거나 의미가 없으면) : break
    tmp = tmp + i번째 값

파이썬이 C 언어가 가지는 문법적 기능을 보다 편하게 사용 가능하도록 손을 봤다는 부분 중 하나라고 볼 수도 있겠지요. 기본적인 for 문이 가지는 구조적 형태와 완전이 일치하는 것을 확인할 수 있습니다. [초기화; 가동 조건; 변화] 라는 구조에서 말이지요.

또한 뒤의 콜론이 생략되는 경우엔 이동 간격이 기본적으로 1으로 선정되도록, 
정지칸이나 출발칸이 생략되는 경우엔 이동 간격으로 갈 수 있는 최대한의 위치가 선정되도록 되어 있습니다.

보통 i = 0으로 놓고 초기화한다던가, 마지막 지점은 끝 값이라는 것을 알고 있을 경우 보다 활용이 쉽게 될 수 있는 구조지요.

아무래도 프로그래밍 언어가 라틴계 영어권 지역에서 만들어지다 보니 생긴 구문적 부분으로, 영어에서 드러나는 시작 -> 핵심 -> 부연 설명 과 같은 구조가 보이는 듯 합니다.

```python
s[3:-3]
```

위에서 고찰했듯이, [3:-3] 은 3의 위치에서 시작해 -3의 위치 이전까지의 문자열을 한 칸씩 이동하며 그 내용을 확인하는 것입니다. 하지만 그것은 사실 제자리걸음과 크게 다르지 않죠. 3의 위치도 -3의 위치도 'h'이기 때문에 이동하는 것에 의미가 없습니다.
위에 for 문의 형태로 표현한 것에서 방향의 결정이 둘 사이의 간격을 절대값으로 나누는 것이라고 한다면 그 간격이 0이 되어 '의미가 없는' 경우에 해당하게 되는 것이지요.
그렇기에 처음 위치를 찍기도 전에 break, 아무 것도 없는 결과가 나타나게 되는 것입니다.

```python
for odd in range(1,100,2):
    print odd,
```

수가 많아져서 지나치게 줄이 길어지는 것을 방지하기 위해 콤마를 덧붙여 출력했습니다. 또한 비교 연산을 거치지 않고 순수하게 홀수만 출력합니다.
range의 초기화 조건은 위에서 고찰한 슬라이싱의 형태와 유사하기 때문에,1로 시작하여 100보다 작은 수를 2씩 더해가며 출력하게 되어, 결과적으로 홀수들만 출력하게 됩니다.

```python
even = 2; sum = 0
while even < 100:
    sum += even
    even += 2
print sum
```

'1과 100 사이'라 할 때 100을 포함하는지 포함하지 않는지의 여부가 다소 불분명하지만, 위에서 계속 따라온 규칙에 맞춰 최종 값에 이르기 전까지인 1 <= even < 100의 범위에서 합을 계산하는 형태로 작성했습니다.
even과 sum을 각각 초기화하고, 해당하는 조건 내에서 값을 계속해서 계산하도록 했습니다.
형태상 for 문보다 메모리의 사용은 적어지게 되겠지만 연산의 수는 늘어나게 됩니다. even의 레퍼런스가 바뀔 때마다 가비지 컬렉터가 수고하는 것까지 생각하면 연산량은 상당히 늘어나게 되겠죠.
for 문으로 작성했던 위 풀이는 range가 임시로 차지하고 있는 공간이 크겠지만 결과적으로 처음에 한 번 하고 마는 방식이니 일장일단이 있을 것입니다.
어느 것이 보다 효율적인지는 사용하는 시스템에 따라 달라질 수 있긴 하겠지만 정밀하면서 최대한 빠른 동작을 원하는 알고리즘 대회 등에선 중요할지도 모릅니다.

```python
import math
num = (int (input("Please Enter a postive Integer") ))
prime = [2]
for nums in range(2, num):
    for primes in prime:
        if(nums % primes ==0):
            break
        elif primes > math.sqrt(nums) and nums not in prime:
            prime.append(nums)
            break
        else:
            continue

print (prime)
```

우선 후에 쓰일 sqrt를 위해서 math 모듈을 임포트하고 시작합니다.
num에 정수 입력을 받되 양수를 받도록 프롬프트합니다. 다만, 문자열로 취급되기 때문에 int로 형 변환을 해줘야 합니다.
그리고 소수를 저장할 리스트를 하나 만들고 for 문을 시작합니다.

2로 시작해 입력받은 수보다 작은 수들에 대해서 작은 반복문을 다시 수행하며, 이 부분이 소수 체크의 핵심 부분입니다.
소수 리스트에 있는 수들로 나눠서 나머지가 없는 경우, 즉 나누어 떨어지는 경우는 소수가 아니므로 바로 break합니다.
만약 현재 확인하는 차례의 소수가 확인하려는 수의 제곱근보다 클 경우, 확인할 만큼 확인했기 때문에 이 수가 소수라는 게 확인됩니다.
또한 여태까지 추가했던 소수 리스트들의 목록을 봐서 중복되지 않도록 넣어 줍니다.
    한 수의 제곱근 이상의 수가 약수인지 확인할 필요는 없습니다. 그럴 경우는 제곱근 이하의 수에서 교환 법칙을 통해 확인되기 때문이죠. 
    가령 10은 그 제곱근인 3.XXX보다 작은 3까지만 확인하면 됩니다. 그보다 큰 약수인 5는 2와 곱해서 10을 만들기 때문에 이미 검증된 것이죠.
만약에 나머지가 있는데 제곱근보다 작을 경우는 넘어갑니다. 이 경우는 우리가 여태까지 찾아왔던 소수들과 중복되며, 이들을 굳이 추가할 필요는 없기 때문입니다. 

위와 같은 절차의 수행 결과는 보시다시피 소수를 잘 걸러낸 걸 확인할 수 있습니다.
